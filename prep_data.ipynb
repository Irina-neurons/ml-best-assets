```python
import pandas as pd
import random

purposes = ['brand_building', 'conversion']



KEY_COLUMNS = ["industry_category", "industry_subcategory", "usecase_category", "usecase_subcategory", "platform", "device", "context"]
BENCHMARK_PATH = "/Users/irinakw/Library/CloudStorage/GoogleDrive-i.white@neuronsinc.com/Shared drives/HQ - R&D/Benchmark Documents/"
METRICS_PATH = "/Users/irinakw/Library/CloudStorage/GoogleDrive-i.white@neuronsinc.com/Shared drives/HQ - R&D/Benchmark Documents/"
```

```python
df_img_level_metrics = pd.read_csv(METRICS_PATH + 'img_nvai_metrics_dec2025.csv')
df_img_aois_metrics = pd.read_csv(METRICS_PATH + 'img_aois_metrics_dec2025.csv')
df_vid_level_metrics = pd.read_csv(METRICS_PATH + 'vid_nvai_metrics_dec2025.csv')
df_vid_aois_metrics = pd.read_csv(METRICS_PATH + 'vid_aois_metrics_dec2025.csv')
df_ab_level_metrics = pd.read_csv(METRICS_PATH + 'ab_nvai_metrics_dec2025.csv')
df_ab_aois_metrics = pd.read_csv(METRICS_PATH + 'ab_aois_metrics_dec2025.csv')
```

```python
df_img_level_metrics = df_img_level_metrics[
    ~df_img_level_metrics['metric'].str.contains('engagement', na=False) | 
    (df_img_level_metrics['metric'] == 'engagement_frt')]
df_img_aois_metrics.aoi_type = df_img_aois_metrics.aoi_type.replace({'call_to_action': 'cta',
                                                                     'body_text': 'bodytext',
                                                               'headline_text': 'headline'})
df_img_aois_metrics['metric'] = df_img_aois_metrics['metric'].replace({
    'aoi_': '',
}, regex=True)
df_img_aois_metrics = df_img_aois_metrics[~df_img_aois_metrics['metric'].str.contains('clarity')]
df_img_aois_metrics = df_img_aois_metrics[~df_img_aois_metrics['metric'].str.contains('engagement')]
df_img_aois_metrics.metric = df_img_aois_metrics.metric + "_" + df_img_aois_metrics.aoi_type
df_img_aois_metrics.drop(columns=['aoi_type','aoi_id'], inplace=True)
df_img = pd.concat([df_img_level_metrics, df_img_aois_metrics], axis=0).sort_values(by='asset_id').reset_index(drop=True)
df_img['asset_type'] = 'image'
df_img
```

```python
# df_vid_level_metrics = df_vid_level_metrics[
#     ~df_vid_level_metrics['metric'].str.contains('engagement', na=False) | 
#     (df_vid_level_metrics['metric'] == 'engagement_frt')]
df_vid_aois_metrics.aoi_type = df_vid_aois_metrics.aoi_type.replace({'call_to_action': 'cta',
                                                                     'body_text': 'bodytext',
                                                               'headline_text': 'headline'})
df_vid_aois_metrics['metric'] = df_vid_aois_metrics['metric'].replace({
    'aoi_': '',
}, regex=True)
df_vid_aois_metrics.metric = df_vid_aois_metrics.metric + "_" + df_vid_aois_metrics.aoi_type
df_vid_aois_metrics.drop(columns=['aoi_type','aoi_id','aoi_name','video_id'], inplace=True)
df_vid = pd.concat([df_vid_level_metrics, df_vid_aois_metrics], axis=0).sort_values(by='asset_id').reset_index(drop=True)
df_vid = df_vid[df_vid.usecase_subcategory != 'animated_banners'].reset_index(drop=True)

df_vid['asset_type'] = 'video'
df_vid
```

```python
# df_ab_level_metrics = df_ab_level_metrics[
#     ~df_ab_level_metrics['metric'].str.contains('engagement', na=False) | 
#     (df_ab_level_metrics['metric'] == 'engagement_frt')]
df_ab_aois_metrics.aoi_type = df_ab_aois_metrics.aoi_type.replace({'call_to_action': 'cta',
                                                                     'body_text': 'bodytext',
                                                               'headline_text': 'headline'})
df_ab_aois_metrics['metric'] = df_ab_aois_metrics['metric'].replace({
    'aoi_': '',
}, regex=True)

df_ab_aois_metrics.metric = df_ab_aois_metrics.metric + "_" + df_ab_aois_metrics.aoi_type
df_ab_aois_metrics.drop(columns=['aoi_type','aoi_id','aoi_name','video_id'], inplace=True)
df_ab = pd.concat([df_ab_level_metrics, df_ab_aois_metrics], axis=0).sort_values(by='asset_id').reset_index(drop=True)

df_ab['asset_type'] = 'animated_banner'
df_ab
```

```python
print("Image df shape:", df_img.shape)
print("Video df shape:", df_vid.shape)
print("Animated Banner df shape:", df_ab.shape)
total_df = pd.concat([df_img, df_vid, df_ab], axis=0).reset_index(drop=True)
print("Total df shape:", total_df.shape)
total_df
```

```python
total_df_pivot = total_df.pivot_table(index=['asset_id', 'asset_type', 'path_bucket'] +KEY_COLUMNS, columns='metric', values='value').reset_index()
print(total_df_pivot.asset_type.value_counts())
total_df_pivot
```
